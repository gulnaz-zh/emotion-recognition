```python
import pandas as pd
import numpy as np
import os
```

```python
path = "ActionUnits/"
elist = os.listdir(path)
```

```python
cnt=0
for csv in elist:
    if cnt==0:
        data = pd.DataFrame(np.load(path+csv))
        cnt+=1
    else:
        d = pd.DataFrame(np.load(path+csv))
        data = pd.concat([data, d])
```
